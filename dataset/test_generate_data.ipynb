```python
import unicodedata

def generer_graine_groupe(nom_complet: str) -> int:
    chaine_normalisee = unicodedata.normalize('NFKD', nom_complet)
    S = "".join(c for c in chaine_normalisee if not unicodedata.combining(c))
    S = S.upper().replace(" ", "")
    
    n = len(S)
    p = 31               
    m = 2**31 - 1        
    graine = 0
    
    for i in range(n):
        valeur_ascii = ord(S[i])
        puissance_p = p ** (n - 1 - i)
        graine = (graine + valeur_ascii * puissance_p) % m
        
    return graine

nom_test = "AVOCAT BEURRE Pierre Le Boulange"
graine_finale = generer_graine_groupe(nom_test)

print(f"Graine numérique obtenue : {graine_finale}")
```

```python
def name_to_seed(nom_complet: str) -> int:
    chaine = nom_complet.upper().replace(" ", "")
    seed = 0
    for caractere in chaine:
        valeur = ord(caractere) - ord('A') + 1  # A=1, ..., Z=26
        seed = (seed * 31 + valeur) % (2**32 - 1)
    return seed

nom_test = "AVOCAT BEURRE Pierre Le Boulange"
graine_finale = name_to_seed(nom_test)

print(f"Graine numérique obtenue : {graine_finale}")
```
